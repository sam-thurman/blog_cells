```python
import yfinance
import pandas as pd
import plotly.express as px
import plotly as pl
```

```python
cos = ['GOOG', 'AAPL', 'AMZN', 'FB', 'NFLX', 'MSFT', 'SPOT', 'TSLA']

closes = []

for co in cos:
    series = yfinance.Ticker(co).history(period='1y').rename({'Close':co}, axis=1)[co]
    closes.append(series)
```

```python
master = pd.concat(closes, axis=1)
master
```

```python
px.line(master.GOOG)
```
