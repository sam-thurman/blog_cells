```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import scipy.stats as st
import random

# np.random.seed(0) #set a random seed for reproducibility
```

```python
non_norm = pd.read_csv('data/non_normal_dataset.csv', squeeze=True)
```

```python
sns.distplot(non_norm)
```

```python
st.normaltest(non_norm)
```

```python
def get_sample(data, n):
    
    return data.sample(n)
```

```python
test_sample = get_sample(non_norm, 30)
print(test_sample[:5]) 
```

```python
def get_sample_mean(sample):
    
    return np.mean(sample)
```

```python
test_sample2 = get_sample(non_norm, 30)
test_sample2_mean = get_sample_mean(test_sample2)
print(test_sample2_mean) 
```

```python
def create_sample_distribution(data, dist_size=100, n=30):
    
    means = []
    while dist_size:
        sample = get_sample(data, n)
        means.append(get_sample_mean(sample))
        dist_size -= 1
    
    return means
```

```python
test_sample_dist = create_sample_distribution(non_norm)
print(test_sample_dist[:5]) 
```

```python
sns.distplot(create_sample_distribution(non_norm, dist_size=10, n=3))
```

```python
sns.distplot(create_sample_distribution(non_norm, dist_size=30, n=10))
```

```python
sns.distplot(create_sample_distribution(non_norm, dist_size=1000, n=30))
```
