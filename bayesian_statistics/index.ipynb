```python
P( beta fish ) = 45/245 -- 45 betas out of a total fish pop of 245
P( small tank ) = 10/245 -- 10 fish in the small tank divided by 245 fish in both tanks
P( beta fish | small tank ) = 1 -- all fish in the small tank are betas
```
