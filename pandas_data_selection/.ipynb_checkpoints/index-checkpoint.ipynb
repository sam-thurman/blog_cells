```python
import pandas as pd 
import numpy as np                 # for numeric operations
import matplotlib.pyplot as plt    # for visualizations
import sklearn.datasets as data    # data we will be using to explore Pandas
```

```python
boston = data.load_boston()
boston
```

```python
X = boston.data
y = boston.target
feat_names = boston.feature_names
feat_descriptions = boston.DESCR

print(X.shape)
print(y.shape)
```

```python
X = pd.DataFrame(X)
X
```

```python
X.columns = feat_names
X
```

```python
df = X
df
```

```python
df['price'] = y
df
```

```python
y = pd.DataFrame(y, columns=['price'])
```

```python
df = pd.concat([X, y], axis=1)
df
```
